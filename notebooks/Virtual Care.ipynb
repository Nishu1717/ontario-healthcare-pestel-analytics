# Linear Trend Regression for Virtual Care

```python

data = {
    'Year': [2019, 2019, 2020, 2020, 2021, 2021, 2022, 2022, 2023, 2023, 2024, 2024],
    'Modality': ['Video', 'Telephone', 'Video', 'Telephone', 'Video', 'Telephone', 'Video', 'Telephone', 'Video', 'Telephone', 'Video', 'Telephone'],
    'Usage_Pct': [9.25, 36.21, 8.07, 37.00, 85.30, 89.90, 68.00, 77.20, 14.25, 28.50, 27.39, 56.47]
}
df = pd.DataFrame(data)

# 2. Prepare for Regression Analysis - We want to calculate the Slope (Growth per Year) for each modality
models = {}
results_text = []

# Set up the plot
plt.figure(figsize=(10, 6))
colors = {'Video': '#E74C3C', 'Telephone': '#2E86C1'} 

for modality in ['Video', 'Telephone']:
    subset = df[df['Modality'] == modality]
    
    X = subset['Year'].values.reshape(-1, 1)
    y = subset['Usage_Pct'].values
    
    # Fit Linear Regression Model
    model = LinearRegression()
    model.fit(X, y)
    
    # Store Slope (Coefficient) and Intercept
    slope = model.coef_[0]
    intercept = model.intercept_
    r_squared = model.score(X, y)
    
    models[modality] = {'Slope': slope, 'R2': r_squared}
    
    results_text.append(f"{modality} Trend: +{slope:.2f}% per year (R²={r_squared:.2f})")
    
    plt.scatter(subset['Year'], subset['Usage_Pct'], color=colors[modality], s=100, zorder=5)
    
    # Plot Regression Line
    # Create a sequence of years for the line (2019 to 2025)
    x_range = np.array(range(2019, 2026)).reshape(-1, 1)
    y_pred = model.predict(x_range)
    plt.plot(x_range, y_pred, color=colors[modality], linestyle='--', label=f"{modality} Trend (Slope: +{slope:.1f})")

    final_val = subset[subset['Year'] == 2024]['Usage_Pct'].values[0]
    plt.text(2024.1, final_val, f"{final_val:.1f}%", color=colors[modality], fontweight='bold')

plt.title("Linear Time-Trend", fontsize=14, fontweight='bold')
plt.ylabel("Population Usage Rate (%)", fontsize=12)
plt.xlabel("Year", fontsize=12)
plt.xticks([2019, 2020, 2021, 2022, 2023, 2024, 2025])
plt.ylim(0, 105)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()

stats_summary = "\n".join(results_text)
plt.text(2020.5, 10, f"Regression Analysis:\n{stats_summary}", 
         bbox=dict(facecolor='white', alpha=0.9, edgecolor='gray'))

plt.tight_layout()
plt.show()

# Print the statistical summary for your report
print("--- REGRESSION RESULTS FOR REPORT ---")
for mod, stats in models.items():
    print(f"{mod}: Slope = {stats['Slope']:.3f} (Growth per year)")
    print(f"{mod}: R-Squared = {stats['R2']:.3f}")
```

# Quadratic time-trend regression

```python
models = {}
results_text = []

plt.figure(figsize=(10, 6))
colors = {"Video": "#E74C3C", "Telephone": "#2E86C1"}  

for modality in ["Video", "Telephone"]:
    subset = df[df["Modality"] == modality].copy()
    
    # centre time at 2019 for numeric stability
    t = (subset["Year"] - 2019).values.reshape(-1, 1)
    X_quad = np.hstack([t, t**2])      
    y = subset["Usage_Pct"].values
    
    model = LinearRegression()
    model.fit(X_quad, y)
    
    r_squared = model.score(X_quad, y)
    models[modality] = {"R2": r_squared}
    results_text.append(f"{modality}: Quadratic R² = {r_squared:.2f}")
    
    # scatter: actual data points
    plt.scatter(
        subset["Year"],
        subset["Usage_Pct"],
        color=colors[modality],
        s=80,
        zorder=3,
        label=f"{modality} (observed)"
    )
    
    # smooth quadratic curve from 2019 to 2024
    years_smooth = np.linspace(2019, 2024, 200)
    t_smooth = (years_smooth - 2019).reshape(-1, 1)
    X_smooth = np.hstack([t_smooth, t_smooth**2])
    y_smooth = model.predict(X_smooth)
    
    plt.plot(
        years_smooth,
        y_smooth,
        linestyle="--",
        color=colors[modality],
        label=f"{modality} trend (quadratic)"
    )

    final_val = subset[subset["Year"] == 2024]["Usage_Pct"].values[0]
    plt.text(2024.05, final_val, f"{final_val:.1f}%", color=colors[modality], fontweight="bold")

plt.title("Quadratic Time-Trend Model", fontsize=14, fontweight="bold")
plt.ylabel("Population Usage Rate (%)", fontsize=12)
plt.xlabel("Year", fontsize=12)
plt.xticks([2019, 2020, 2021, 2022, 2023, 2024])
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend()

stats_summary = "\n".join(results_text)
plt.text(
    2019.2,            
    df["Usage_Pct"].min() + 3,   
    f"Trend model (quadratic):\n{stats_summary}",
    bbox=dict(facecolor="white", alpha=0.9, edgecolor="gray")
)

plt.tight_layout()
plt.show()

# print a clean summary for the report
print("--- QUADRATIC TREND RESULTS FOR REPORT ---")
for mod, stats in models.items():
    print(f"{mod}: R-Squared = {stats['R2']:.3f}")
```
